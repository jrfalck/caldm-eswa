```python
import sys
import subprocess
import pandas as pd
import numpy as np
from pathlib import Path

# Adjust REPO_ROOT if your notebook is not inside a notebooks/ subfolder
REPO_ROOT   = Path("..").resolve()
SCRIPTS_DIR = REPO_ROOT / "scripts"
RESULTS_DIR = REPO_ROOT / "results"
DATA_PATH   = REPO_ROOT / "data" / "final_data_subsetFLGAAL8.npz"

print("Repo root   :", REPO_ROOT)
print("Scripts dir :", SCRIPTS_DIR)
print("Results dir :", RESULTS_DIR)
print("Data file   :", DATA_PATH, "| exists:", DATA_PATH.exists())
```

```python
# Set RUN_TRAINING = True  to re-run the full ablation from scratch
# Set RUN_TRAINING = False to load pre-computed results (default, no GPU needed)
#
# WARNING: RUN_TRAINING = True requires an NVIDIA GPU and approximately
# 4 hours of compute time (6 configurations x 5 seeds x 45 training epochs each).

RUN_TRAINING = True

# Only used when RUN_TRAINING = True
SEEDS      = [7, 42, 888, 1337, 2024]
OUTPUT_DIR = str(RESULTS_DIR)

if RUN_TRAINING:
    if not DATA_PATH.exists():
        raise FileNotFoundError(
            f"\nBLS dataset not found at:\n  {DATA_PATH}\n\n"
            "Please download the dataset from:\n"
            "  https://data.hrmatica.com/wagescale\n"
            "and place the .npz file at the path above."
        )

    import sys
    sys.path.insert(0, str(SCRIPTS_DIR))
    from caldm_ablation_study_v3 import run_ablation

    print("Running ablation study...")
    print("Seeds:", SEEDS)
    print("Estimated time: approximately 4 hours on an NVIDIA RTX-class GPU.\n")

    run_ablation(
        data_path=str(DATA_PATH),
        seeds=SEEDS,
        output_dir=str(RESULTS_DIR),
    )

    print("\nAblation complete. Results saved to:", RESULTS_DIR)

else:
    print("RUN_TRAINING = False — loading pre-computed results.")
```

```python
summary_path  = RESULTS_DIR / "ablation_results_summary.csv"
per_type_path = RESULTS_DIR / "ablation_per_type.csv"

if not summary_path.exists():
    raise FileNotFoundError(
        f"Results not found at {summary_path}.\n"
        "Either set RUN_TRAINING = True in Cell 2, or ensure the pre-computed\n"
        "results CSVs are present in the results/ folder."
    )

summary  = pd.read_csv(summary_path)
per_type = pd.read_csv(per_type_path)

print(f"Loaded summary  : {len(summary)} rows  (expect 6 — one per configuration)")
print(f"Loaded per-type : {len(per_type)} rows  (expect 6 — one per configuration)")
print()
print("Configurations found:")
for _, r in summary.iterrows():
    print(f"  {r['config']:<25}  {r['label']}")
```

```python
LABEL_MAP = {
    "VAE only":                          "A. VAE-Only",
    "Data-space DDPM":                   "B. Data-Space DDPM",
    "Latent Diffusion (unconditional)":  "C. Latent Diff (Uncond.)",
    "Context in VAE only":               "D. Cxt-VAE Only",
    "CALDM (proposed)":                  "E. CALDM-D (proposed)",
    "CALDM + VAE context":               "F. CALDM-J (jointly conditioned)",
}

def fmt(mean, std, decimals=3):
    return f"{mean:.{decimals}f} \u00b1 {std:.{decimals}f}"

def fmt_pval(val):
    if pd.isna(val):   return "\u2014"
    if val < 0.0001:   return "<0.0001"
    if val < 0.001:    return f"{val:.4f}"
    return f"{val:.3f}"

rows = []
for _, r in summary.iterrows():
    label = r["label"]
    # Use 5 decimal places for F1 on C and E — near-zero variance, 3dp gives identical values
    f1_dec = 5 if label in ("Latent Diffusion (unconditional)", "CALDM (proposed)") else 3
    rows.append({
        "Configuration": LABEL_MAP.get(label, label),
        "AUROC":         fmt(r["auc_roc_mean"], r["auc_roc_std"]),
        "AP":            fmt(r["ap_mean"],       r["ap_std"]),
        "F1":            fmt(r["f1_mean"],       r["f1_std"], decimals=f1_dec),
        "p (AUROC)":     fmt_pval(r.get("auc_roc_pval_vs_full")),
        "p (AP)":        fmt_pval(r.get("ap_pval_vs_full")),
    })

table7 = pd.DataFrame(rows)
print("Table 7. Ablation Study Results \u2014 BLS Dataset")
print("(mean \u00b1 std across 5 random seeds; p-values vs CALDM-D)\n")
print(table7.to_string(index=False))


# ============================================================
# CELL 6 — Reproduce Table 8 (per-outlier-type AUROC)
# ============================================================

# per_type columns: config, label,
#   auc_type_HRS_mean, auc_type_HRS_std,
#   auc_type_INA_mean, auc_type_INA_std, ... etc.

TYPES = ["HRS", "INA", "LRS", "PER", "PVO"]

rows = []
for _, r in per_type.iterrows():
    row = {"Configuration": LABEL_MAP.get(r["label"], r["label"])}
    for ot in TYPES:
        row[ot] = fmt(r[f"auc_type_{ot}_mean"], r[f"auc_type_{ot}_std"])
    rows.append(row)

table8 = pd.DataFrame(rows)
print("Table 8. Per-Outlier-Type AUROC \u2014 BLS Dataset")
print("(mean \u00b1 std across 5 seeds; proposed CALDM-D row = E)\n")
print(table8.to_string(index=False))
```

```python
# ============================================================
# CELL 7 — Verify key statistical claims from the paper
# ============================================================

e    = summary[summary["label"] == "CALDM (proposed)"].iloc[0]
c    = summary[summary["label"] == "Latent Diffusion (unconditional)"].iloc[0]
b    = summary[summary["label"] == "Data-space DDPM"].iloc[0]
d    = summary[summary["label"] == "Context in VAE only"].iloc[0]
f    = summary[summary["label"] == "CALDM + VAE context"].iloc[0]
e_pt = per_type[per_type["label"] == "CALDM (proposed)"].iloc[0]
c_pt = per_type[per_type["label"] == "Latent Diffusion (unconditional)"].iloc[0]
d_pt = per_type[per_type["label"] == "Context in VAE only"].iloc[0]
f_pt = per_type[per_type["label"] == "CALDM + VAE context"].iloc[0]

print("=" * 62)
print("Section 7.1 — Aggregate performance claims")
print("=" * 62)
print(f"\nCALDM-D vs Config C (unconditional latent diffusion):")
print(f"  \u0394AUROC = {e['auc_roc_mean'] - c['auc_roc_mean']:+.3f}   paper: +0.020")
print(f"  \u0394AP    = {e['ap_mean']      - c['ap_mean']:+.3f}   paper: +0.026")
print(f"  p(AUROC) = {c['auc_roc_pval_vs_full']:.3f}   paper: 0.008")
print(f"  p(AP)    = {c['ap_pval_vs_full']:.3f}   paper: 0.014")

print(f"\nCALDM-D AUROC : {e['auc_roc_mean']:.3f} \u00b1 {e['auc_roc_std']:.3f}   paper: 0.748 \u00b1 0.007")
print(f"Config C AUROC: {c['auc_roc_mean']:.3f} \u00b1 {c['auc_roc_std']:.3f}   paper: 0.729 \u00b1 0.002")

gap_bc = (c["auc_roc_mean"] - b["auc_roc_mean"]) * 100
print(f"\n22-point gap (B vs C): {gap_bc:.1f} pp   paper: 22")

print("\n" + "=" * 62)
print("Section 7.2 — Stability / variance claims")
print("=" * 62)

joint_mean_ap_std  = (d["ap_std"] + f["ap_std"]) / 2
caldm_d_ap_std     = e["ap_std"]
nojoint_ap_stds    = summary[summary["label"].isin([
    "VAE only", "Data-space DDPM",
    "Latent Diffusion (unconditional)", "CALDM (proposed)"])]["ap_std"]
min_nojoint_ap_std = nojoint_ap_stds.min()

print(f"\nJoint (D, F) mean AP std  : {joint_mean_ap_std:.4f}")
print(f"CALDM-D (E) AP std        : {caldm_d_ap_std:.4f}")
print(f"Min non-joint AP std (C)  : {min_nojoint_ap_std:.4f}")
print(f"\nRatio joint / CALDM-D       : {joint_mean_ap_std / caldm_d_ap_std:.1f}x   paper: ~4x")
print(f"Ratio joint / min non-joint : {joint_mean_ap_std / min_nojoint_ap_std:.1f}x   paper: up to 47x")

print("\n" + "=" * 62)
print("Section 7.3 — Per-type characterization claims")
print("=" * 62)

lrs_delta = e_pt["auc_type_LRS_mean"] - c_pt["auc_type_LRS_mean"]
print(f"\nLRS: CALDM-D = {e_pt['auc_type_LRS_mean']:.3f}   paper: 0.762")
print(f"LRS: Config C = {c_pt['auc_type_LRS_mean']:.3f}   paper: 0.623")
print(f"LRS advantage = {lrs_delta:+.3f}   paper: +0.139")

print(f"\nPER: Config D = {d_pt['auc_type_PER_mean']:.3f}   paper: 0.596")
print(f"PER: Config F = {f_pt['auc_type_PER_mean']:.3f}   paper: 0.596")
print(f"PER: CALDM-D  = {e_pt['auc_type_PER_mean']:.3f}   paper: 0.532")

print(f"\nINA: Config C = {c_pt['auc_type_INA_mean']:.3f}   paper: 0.983")
print(f"INA: CALDM-D  = {e_pt['auc_type_INA_mean']:.3f}   paper: 0.977")

print(f"\nHRS: Config C = {c_pt['auc_type_HRS_mean']:.3f}   paper: 0.694")
print(f"PVO: Config C = {c_pt['auc_type_PVO_mean']:.3f}   paper: 0.838")
```
